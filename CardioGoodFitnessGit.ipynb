```python


# Age distribution
plt.figure(figsize=(10, 6))
df['Age'].plot(kind='hist', bins=15, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Product usage by gender
plt.figure(figsize=(10, 6))
usage_by_gender = df.groupby(['Product', 'Gender'])['Usage'].mean().unstack()
usage_by_gender.plot(kind='bar', figsize=(10, 6))
plt.title('Average Product Usage by Gender')
plt.xlabel('Product')
plt.ylabel('Average Usage')
plt.grid(True)
plt.show()
```

```python
# Income distribution by product
plt.figure(figsize=(10, 6))
df.boxplot(column='Income', by='Product', grid=False)
plt.title('Income Distribution by Product')
plt.suptitle('')  # Suppress the default title
plt.xlabel('Product')
plt.ylabel('Income')
plt.grid(True)
plt.show()
```

```python
# Miles vs. Income scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Income'], df['Miles'], alpha=0.5)
plt.title('Miles vs. Income')
plt.xlabel('Income')
plt.ylabel('Miles')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Fitness'], df['Miles'], alpha=0.5)
plt.title('Miles vs. Fitness')
plt.xlabel('Fitness')
plt.ylabel('Miles')
plt.grid(True)
plt.show()
```

```python
# Count of purchases for each product
product_counts = df['Product'].value_counts()

# Plotting the purchase quantities for the 3 products
plt.figure(figsize=(10, 6))
product_counts.plot(kind='bar', edgecolor='black')
plt.title('Purchase Quantities of the 3 Products')
plt.xlabel('Product')
plt.ylabel('Quantity Purchased')
plt.grid(axis='y')

plt.show()
```
